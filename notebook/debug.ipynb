```python
from pathlib import Path

import numpy as np

input_dir = Path('/home/kuto/kaggle/kaggle-sleep-v2/output/processed_data/train/')

length_list = []
for file_path in input_dir.glob('*/duplicate_len.npy'):
    length_list.append(np.load(file_path))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
plt.hist(np.concatenate(length_list),bins=100)
```

```python

np.concatenate(length_list).mean(), np.concatenate(length_list).std(ddof=1)
```
